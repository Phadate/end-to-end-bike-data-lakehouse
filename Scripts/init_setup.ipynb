## CREATE CATALOG

```sql
CREATE CATALOG IF NOT EXISTS bike_lakehouse
COMMENT 'This is a catalog for the bike lakehouse accelerator' 
```

```sql
USE CATALOG bike_lakehouse;
```

## CREATE SCHEMAS

```sql
CREATE SCHEMA IF NOT EXISTS bronze
COMMENT 'Raw data';

CREATE SCHEMA IF NOT EXISTS silver
COMMENT 'Cleaned data and transforemed';

CREATE SCHEMA IF NOT EXISTS gold
COMMENT 'Aggregated data and businss ready';
```
