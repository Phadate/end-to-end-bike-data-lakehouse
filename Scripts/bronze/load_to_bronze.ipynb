# LOAD DATA FROM SOURCE VOLUME INTO THE BROZE LAYER 

## IMPORT THE INGESTION DATA AND INTITIATE THE CATALOG AND SCHEMA

```python
from bronze_config import INGESTION_DATA
```

```python
%sql
USE CATALOG bike_lakehouse;
USE SCHEMA bronze;
```

### INGEST DATA FROM SOURCE to BRONZE TABLE

```python
for item in INGESTION_DATA:
    print(f"Ingesting file from {item['source']} into bronze table {item['table_name']}")

    df = (
        spark.read
        .option("header", "true")
        .option("inferSchema", "true")
        .option("sep", ",")
        .csv(item["path"])
    )

    (
        df.write
        .format("delta")
        .mode("overwrite")
        .saveAsTable(item["table_name"])
    )
print("Data ingestion completed.")
```
